```python
# creating the array that the python script will use -> starting with 1 billion

data_array = []
for i in range(1_000_000_000):
    data_array.append(i + 1)

print(f"Array of integers created {data_array[-1]}")
```

```python
# will search one item at a time 

def linear_search(target, array):
    sorted_array = sorted(array)
    
    for index, item in enumerate(sorted_array):
        if item == target:
            return index, target
        elif index + 1 <= len(sorted_array):
            index += 1
    return print("file not found")
        

```

```python
# log2(x) ---> 

def binary_search(target, array):
    left, right = 0, len(array) - 1 #left is 0 and right is the len of the array
    
    while left <= right:
        mid = left + (right - left) // 2

        if array[mid] == target:
            return mid
        elif array[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1  # Target not found
```

```python
linear_search(1_000_000_000, data_array)
```

```python
binary_search(1_000_000_000, data_array)

```
