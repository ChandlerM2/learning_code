```python
# creating the array that the python script will use -> starting with 1 billion

data_array = []
for i in range(100_000_000):
    data_array.append(i + 1)

print(f"Array of integers created {data_array[-1]}")
```

```python
# will search one item at a time 

def linear_search(target, array):
    sorted_array = sorted(array)
    
    for index, item in enumerate(sorted_array):
        if item == target:
            return index, target
        elif index + 1 <= len(sorted_array):
            index += 1
    return print("file not found")
        

```

```python
# log2(x) ---> must remember this is working with the index not the values 

def binary_search(target, array):
    low = 0 # index valye
    high = len(array) - 1 #O(1), finding end index value

    while low <= high: # when statement is false we exit
        mid = (low + high) // 2 
        guess = array[mid]

        if guess == target:
            return mid
        elif guess > target:
            high = mid - 1
        else: 
            low = mid + 1
    return print("file not found")





```
