```python
# creating the array that the python script will use

data_array = []
for i in range(1_000_000):
    data_array.append(i + 1)

print(f"Array of 1,000,000 integers created {data_array[-1]}")
```

```python
def linear_search(target, array):
    sorted_array = sorted(array)
    index = 0
    for index, item in enumerate(sorted_array):
        if item == target:
            return index, target
        elif index <= len(sorted_array):
            index += 1
        else:
            return "error: value not found"

linear_search(100_000, data_array)
```

```python
linear_search(8000, data_array)
```
